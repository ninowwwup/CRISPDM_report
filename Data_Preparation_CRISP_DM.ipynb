# 1º Dataframe **'DataPrepFinal.csv'**

```python
# Importando e lendo o 1º dataframe 'DataPrepFinal'

df = pd.read_csv("DataPrepFinal.csv")
```

```python
# Visualizando o df (dataframe)
df.head()
```

```python
# Dropando (Retirando) do df colunas que não serão necessárias às próximas análises

df = df.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
```

```python
/# Visualizando o novo df limpado

df.head()
```

```python
# Visualizando a tipagem das colunas com .dtypes
df.dtypes
```

```python
'''Utilizando uma função de filtro (lambada) para encontrar dentro da 
coluna 'usd_goal_real', valores do tipo strings dropando esse valor com .apply, 
guardando esses novos valores dentro da mesma coluna 'usd_goal_real'.
'''
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
df.head()
```

Link para explicação de REGEX:  https://www.w3schools.com/python/python_regex.asp

Link para testar REGEX: https://www.regextester.com/

```python
''' 1. Utilizando a fução regex=True para acessar valores dentro da culona
'launched', e mudando esses valores do tipo str com o .replace, salvando 
dentro da mesma coluna.
'''

df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
# Visualizando 

df.head()
```

```python
# Verificando o tipo

df.dtypes
```

```python
# Mudadando a culuna 'usd_goal_real' do tipo object(str) para o tipo int(int64) com a função .astype

df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
```

```python
# Verificando o tipo

df.dtypes
```

```python
# Verficando as colunas 
df.head()
```

Link do tipes de formatos datetime do pandas:https://www.dataindependent.com/pandas/pandas-to-datetime/

```python
# Trandformando o tipo de data de duas colunas que estão com a tiapagem object(str) para o formato de datas(datetime64[ns])

df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```

```python
# Verificando o como está ficando as colunas

df.dtypes
```

```python
# Criando uma nova coluna 'time_range' dentro do dataframe df com a informação do tempo que a campanha ficou aberta
# Pra isso pegamos a coluna de ínicio 'deadline' menos(-) a coluna de lançamento 'launched'

df['time_range'] = (df['deadline']-df['launched']).dt.days
```

```python
df.head()
```

## One Hot Encoding

### É uma técnica utilizada em python, apropriadamente em ML, essa técnica consiste em transformar uma variável do tipo CATEGÓRICA em variável tipo BOLEANO

```python
# Pegado a coluna categórica 'backedLocation' e Transformando-a em boleanos
# Guardando em um novo df_invest

df_invest = pd.get_dummies(df_invest, columns= ['backedLocation'])

'''Esse código acessou a coluna 'backedLocation' pegando as categorias dentro dessa coluna (BR, GBK, US)
e transfornou as informações contidas dentro delas em valores bináros (boleanos)
'''
```

```python
# Verificando a quandtidade de investimento(Registro) para cada ID

df_invest.sort_values(by=['ID'])
```

```python
# Agrupando os dados por ID e agregando os valores (.agg)
# Tirando a média da coluna de idade 'age'
# Somando os dados dentro das colunas 'backedLocation_BR', 'backedLocation_GBK', 'backedLocation_US'
# Atualizando o dadaframe com os novos dados df_invest

df_invest = df_invest.groupby(by=['ID']).agg({'age':'mean', 'backedLocation_BR': 'sum', 'backedLocation_GBK': 'sum', 'backedLocation_US': 'sum'})
```

```python
# Verificando 

df_invest.head()
```

```python
# Unindo as as duas tabelas (df e df_invest) com o (.merge) pelo ID (on='ID'). Pela direita utilizando o parâmetro (how='right)
# Salvando em uma nova variável df

df =  df.merge(df_invest, how='right', on='ID')
```
